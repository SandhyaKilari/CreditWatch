## Data Preprocessing

### Rename dependent variable

```python
df['IsDefaulter'] =df ['default.payment.next.month']
df = df.drop('default.payment.next.month',axis = 1)
df
```

### Data Cleaning

```python
columns_to_check = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'IsDefaulter']

unique_values = {col: df[col].unique() for col in columns_to_check}

print(unique_values)
```

```python
df['MARRIAGE'].unique()
```

```python
df['MARRIAGE'] = df['MARRIAGE'].replace(0, 3)
df = df[df['MARRIAGE'] != 3]
df['MARRIAGE'].unique()
```

```python
df['EDUCATION'].unique()
```

```python
df['EDUCATION'] = df['EDUCATION'].apply(lambda x: 4 if x in [0, 5, 6] else x)
df = df[df['EDUCATION'] != 4]
df['EDUCATION'].unique()
```

```python
df
```

### Rename few independent column varaibles

```python
df.rename(columns={'PAY_0':'PAY_SEPT','PAY_2':'PAY_AUG','PAY_3':'PAY_JUL','PAY_4':'PAY_JUN','PAY_5':'PAY_MAY','PAY_6':'PAY_APR'},inplace=True)
df.rename(columns={'BILL_AMT1':'BILL_AMT_SEPT','BILL_AMT2':'BILL_AMT_AUG','BILL_AMT3':'BILL_AMT_JUL','BILL_AMT4':'BILL_AMT_JUN','BILL_AMT5':'BILL_AMT_MAY','BILL_AMT6':'BILL_AMT_APR'}, inplace = True)
df.rename(columns={'PAY_AMT1':'PAY_AMT_SEPT','PAY_AMT2':'PAY_AMT_AUG','PAY_AMT3':'PAY_AMT_JUL','PAY_AMT4':'PAY_AMT_JUN','PAY_AMT5':'PAY_AMT_MAY','PAY_AMT6':'PAY_AMT_APR'},inplace=True)
```

```python
df
```

### Drop ID column

```python
df = df.drop('ID', axis=1)
```

```python
df
```

### Class imbalance: Oversampling

```python
category_counts = df['IsDefaulter'].value_counts()
print(category_counts)
```

```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)
X_over, y_over = oversampler.fit_resample(df.drop('IsDefaulter', axis=1), df['IsDefaulter'])
df = pd.concat([X_over, y_over], axis=1)
df['IsDefaulter'].value_counts()
```

### Feature Engineering

```python
df_final = df.copy()
```

```python
df_final['Dues'] = (df_final['BILL_AMT_APR']+df_final['BILL_AMT_MAY']+df_final['BILL_AMT_JUN']+df_final['BILL_AMT_JUL']+df_final['BILL_AMT_SEPT'])-(df_final['PAY_AMT_APR']+df_final['PAY_AMT_MAY']+df_final['PAY_AMT_JUN']+df_final['PAY_AMT_JUL']+df_final['PAY_AMT_AUG']+df_final['PAY_AMT_SEPT'])
```

```python
df_final = pd.get_dummies(df_final, columns = ['PAY_SEPT', 'PAY_AUG', 'PAY_JUL', 'PAY_JUN', 'PAY_MAY', 'PAY_APR'], drop_first = True )
```

```python
column_names = {
    'SEX_1': 'Male',
    'SEX_2': 'Female',
    'EDUCATION_1': 'Graduate School',
    'EDUCATION_2': 'University',
    'EDUCATION_3': 'High School',
    'MARRIAGE_1': 'Married',
    'MARRIAGE_2': 'Single'
}

df_final = df_final.rename(columns=column_names).astype(int)
df_final
```

```python
df_final.columns
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

X = df_final.drop('IsDefaulter', axis=1) 
y = df_final['IsDefaulter']


scaler = StandardScaler()
for column in X.columns:
    X[column] = scaler.fit_transform(X[[column]])
    
```

### Outliers in the Dataset

```python
import numpy as np
from scipy import stats  

numerical_cols = [
    'LIMIT_BAL', 'BILL_AMT_SEPT', 'BILL_AMT_AUG', 'BILL_AMT_JUL', 'BILL_AMT_JUN', 'BILL_AMT_MAY', 'BILL_AMT_APR', 'PAY_AMT_SEPT',
       'PAY_AMT_AUG', 'PAY_AMT_JUL', 'PAY_AMT_JUN', 'PAY_AMT_MAY', 'PAY_AMT_APR'
]

data_zscore = X[numerical_cols].apply(stats.zscore)

threshold = 3

outliers = np.where(data_zscore > threshold)

X = X[(data_zscore < threshold).all(axis=1)]
y = y[(data_zscore < threshold).all(axis=1)]
```

## Feature Selection/ Dimensionality Reduction

```python
## Principal Component Analysis (PCA)

import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X)

explained_variance_ratio = pca.explained_variance_ratio_

cumulative_variance = np.cumsum(explained_variance_ratio)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center', label='Explained Variance Ratio')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio / Cumulative Explained Variance')
plt.title('Explained Variance Ratio and Cumulative Explained Variance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

pca = PCA(n_components=8)
X_pca = pca.fit_transform(X)

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8'])
pca_df['Default'] = y.map({0: 'Not Defaulter', 1: 'Defaulter'})  # Assigning meaningful labels to hue elements


# Plot the scatter plot matrix of all principal components
sns.pairplot(pca_df, hue='Default', diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Scatter Plot Matrix of Principal Components')
plt.show()
```
